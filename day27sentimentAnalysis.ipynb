```python
import re
```

```python
def clean_text(text):
    text=re.sub(r'\W',' ',text)
    text=re.sub(r'\d+',' ',text)
    text=re.sub(r'\s',' ',text)
    return text.lower()
```

```python
df['text']=df['text'].apply(clean_text)
```

```python
df.head()
```

```python
#step1 : removing the stopwords
#step2 : lemmatizing
#step3 : vectorizing
```

```python
import nltk
```

```python
from nltk.corpus import stopwords
```

```python
stopwords=set(stopwords.words('english'))
```

```python
df['text']=df['text'].str.split()
```

```python
stopwords
```

```python
df['text']=df['text'].apply(lambda x:[word for word in x if word not in stopwords])
```

```python
df
```

```python
#lemmatizing
```

```python
from nltk.stem import PorterStemmer
```

```python
stem=PorterStemmer()
```

```python
df['text stemmed']=df.text.apply(lambda x:[stem.stem(word) for word in x])
```

```python
df
```

```python
df['text stemmed']=df['text stemmed'].apply(lambda x:' '.join(x))
```

```python
import sklearn
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
cv=TfidfVectorizer(max_features=5000)
```

```python
x=cv.fit_transform(df['text stemmed']).toarray()
```

```python
y=df['sentiment']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model= MultinomialNB()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
#finding errors
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mean_squared_error(y_pred,y_test)
```

```python
r2_score(y_pred,y_test)
```
