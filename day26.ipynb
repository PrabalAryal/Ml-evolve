```python
def clean_text(text):
    word=[]
    for i in ast.literal_eval(text):
        word.append(i)
    return word
```

```python
data2['tags']=data2['tags'].apply(clean_text)
```

```python
data2
```

```python
import nltk
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
stop_words=set(stopwords.words('english'))
```

```python
data2.tags=data2.tags.apply(lambda x:[word for word in x if word not in stop_words])
```

```python
data2.tags=data2.tags.apply(lambda x:' ' .join(x))
```

```python
data2.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
data2
```

```python
import sklearn
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000)
```

```python
vector=cv.fit_transform(data2['tags']).toarray()
```

```python
vector
```

cosine similarity 

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend_movie(movie):
    index=data2[data2['title']==movie].index[0]
    distances=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])
    for i in distances[1:6]:
        print(data2.iloc[i[0]].title)
```

```python
recommend_movie('The Dark Knight Rises')
```
