```python
#find which occupation person  shopped based on amount
data[data['Amount'].max()==data['Amount']][['Occupation','Amount']]
```

```python
#maximum amount spent based on gender and state 
data.groupby(['State','Gender'])['Amount'].sum().reset_index()
```

```python
data.groupby(['Gender','Product_Category'])['Amount'].sum().reset_index().sort_values(ascending=False,by='Amount')
```

```python
#count plot with number of males and females shown
ax=sns.countplot(x='Gender',data=data,palette='cool',hue='Amount')
for bars in ax.containers:
    ax.bar_label(bars)
```
