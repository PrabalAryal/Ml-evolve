```python
def capping_outliers(data):
    q1=data.quantile(0.25)
    q3=data.quantile(0.75)
    iqr=q3-q1
    upper_limit=q3+1.5*iqr
    lower_limit=q1-1.5*iqr
    data.loc[data > upper_limit]=upper_limit
    data.loc[data < lower_limit]=lower_limit
    return data

cols=['Amount Requested','Amount Funded By Investors','Monthly Income','Revolving CREDIT Balance']
for col in cols:
    cap1_data[col]=capping_outliers(cap1_data[col])
```

```python
X_scaled_train,X_scaled_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.2,random_state=42)
```
